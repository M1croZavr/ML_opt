```python
from coreml.classification import models
```

```python
from sklearn import datasets, metrics, linear_model, model_selection, svm, preprocessing
# Linear separatable
X, y = datasets.make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, class_sep=1.5)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)
print('Linear class balance:', sum(y_test == 0), sum(y_test == 1))
# Circles
Xc, yc = datasets.make_circles(n_samples=500, noise=0.05)
Xc_train, Xc_test, yc_train, yc_test = model_selection.train_test_split(Xc, yc, test_size=0.2)
print('Circles class balance:', sum(yc_test == 0), sum(yc_test == 1))
```

## LogisticRegressionRidge

```python
lr = float(input('Enter learning rate (float number), default=0.001: '))
eps = float(input('Enter convergence condition coefficient (float number), default=0.00001: '))
C = float(input('Enter inverse of regularization coefficient (float number), default=1: '))
if int(input('Do you want to use polynomial? 1 - Yes, 0 - No: ')):
    degree = int(input('Enter maximum degree of generating polynomial features (int number): '))
else:
    degree = None
plot = bool(int(input('Should we plot the result, 1 - Yes, 0 - No: ')))

lr_ridge = models.LogisticRegressionRidge(C=C, eps=eps, lr=lr, degree=degree, plot=plot)
lr_ridge.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_ridge.predict(X_test), y_test))
```

## LogisticRegressionRBF

```python
C = float(input('Enter inverse of regularization coefficient (float number), default=1: '))
eps = float(input('Enter convergence condition coefficient (float number), default=0.00001: '))
gamma = float(input('Enter rbf hyperparameter for exp(-gamma||x - l||^2), (float number), default=0.1: '))
plot = bool(int(input('Should we plot the result, 1 - Yes, 0 - No: ')))

lr_rbf = models.LogisticRegressionRBF(C=C, eps=eps, gamma=gamma, plot=plot)
lr_rbf.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_rbf.predict(X_test), y_test))
```

## LogisticRegressionLasso

```python
lr = float(input('Enter learning rate (float number), default=0.001: '))
eps = float(input('Enter convergence condition coefficient (float number), default=0.00005: '))
C = float(input('Enter inverse of regularization coefficient (float number), default=1: '))
if int(input('Do you want to use polynomial? 1 - Yes, 0 - No: ')):
    degree = int(input('Enter maximum degree of generating polynomial features (int number): '))
else:
    degree = None
plot = bool(int(input('Should we plot the result, 1 - Yes, 0 - No: ')))

lr_lasso = models.LogisticRegressionLasso(C=C, eps=eps, lr=lr, degree=degree, plot=plot)
lr_lasso.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_lasso.predict(X_test), y_test))
```

## SupportVectorClassifier

```python
C = float(input('Enter inverse of regularization coefficient (float number), default=1: '))
kernel = input("Specify the kernel type to be used in the algorithm. {'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'}, default='rbf': ")
eps = float(input('Enter convergence condition coefficient (float number), default=0.001: '))
degree = int(input('Enter degree of the polynomial kernel function (int number), default=3: '))
plot = bool(int(input('Should we plot the result, 1 - Yes, 0 - No: ')))

linear_svc = models.SupportVectorClassifier(C=C, eps=eps, kernel=kernel, degree=degree, plot=plot)
linear_svc.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(linear_svc.predict(X_test), y_test))
```

## SupportVectorClassifierDual

```python
reg = input("Enter Which regularization to use. {'ridge', 'lasso'}, default='ridge': ")
C = float(input('Enter inverse of regularization coefficient (float number), default=1: '))
eps = float(input('Enter convergence condition coefficient (float number), default=0.005: '))
if int(input('Do you want to use polynomial? 1 - Yes, 0 - No: ')):
    degree = int(input('Enter maximum degree of generating polynomial features (int number): '))
else:
    degree = None
plot = bool(int(input('Should we plot the result, 1 - Yes, 0 - No: ')))

dual_svc = models.SupportVectorClassifierDual(reg=reg, C=C, eps=eps, degree=degree, plot=plot)
dual_svc.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(dual_svc.predict(X_test), y_test))
```
