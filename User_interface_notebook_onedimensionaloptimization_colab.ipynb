## Метод золотого сечения

```python
params_names = ['fun_anl',
                'x_l',
                'x_r',
                'eps',
                'max_iter',
                'print_info',
                'record_info',
                'plot']
parameters = {}
for param_name in params_names:
    if param_name == 'fun_anl':
        param_value = input('Enter function: ')
        parameters[param_name] = param_value
    elif param_name in params_names[1:3]:
        param_value = float(input(param_name + ' required: '))
        parameters[param_name] = param_value
    else:
        param_value = eval(input(param_name + ' (False if leave default: '))
        if param_value:
            parameters[param_name] = param_value
optimize.golden_section_search(**parameters)
```

```python
f_analytic = input('Enter the function in analytical form: f = ')
x_l = float(input('Enter the left bound of the interval: '))
x_r = float(input('Enter the right bound of the interval: '))
eps = int(input('Change precision of result(default=1e-5), 1 - change, 0 - use default: '))
if eps:
    eps = float(input('Enter the right bound of the interval'))
else:
    eps = 1e-5
max_iter = int(input('Change maximum iteration of algorithm(default=500), 1 - change, 0 - use default: '))
if max_iter:
    int(input('Enter the right bound of the interval'))
else:
    max_iter = 500
print_info = bool(int(input('Should we print information in each iteration , 1 - yes, 0 - no: ')))
record_info = bool(int(input('Make pd.DataFrame with recorder information , 1 - yes, 0 - no: ')))
plot = bool(int(input('Should we draw a plot , 1 - yes, 0 - no: ')))
if record_info:
    f_res, x_res, df = optimize.golden_section_search(f_analytic, x_l, x_r, eps, max_iter, print_info, record_info, plot)
    print('f_analytic:', f_analytic,'x_res:', x_res,'\nres in DataFrame\n', df)
else:
    f_res, x_res = optimize.golden_section_search(f_analytic, x_l, x_r, eps, max_iter, print_info, record_info, plot)
    print('f_analytic:', f_analytic,'x_res:', x_res)
```

## Алгоритм Бройдена — Флетчера — Гольдфарба — Шанно

```python
params_names = ['fun_anl',
                'x0',
                'c1',
                'c2',
                'x_max',
                'threshold',
                'max_iter',
                'nu',
                'print_info',
                'record_info',
                'plot']
parameters = {}
for param_name in params_names:
    if param_name == 'fun_anl':
        param_value = input('Enter function: ')
        parameters[param_name] = param_value
    elif param_name == 'x0':
        param_value = float(input(param_name + ' required: '))
        parameters[param_name] = param_value
    else:
        param_value = eval(input(param_name + ' (False if leave default: '))
        if param_value:
            parameters[param_name] = param_value
optimize.bfgs(**parameters)
```

```python
f_analytic = input('Enter the function in analytical form: f = ')
x0 = int(input('Enter starting point: '))
c1 = int(input('Should we change the first constant of the Wolfe condition(default=0.0001), 1 - change, 0 - use default: '))
if c1:
    c1 = float(input('Enter the right bound of the interval'))
else:
    c1=0.0001
c2 = int(input('Should we change second constant of the Wolfe condition(default=0.1), 1 - change, 0 - use default: '))
if c2:
    c2 = float(input('Enter the right bound of the interval'))
else:
    c2=0.1
x_max = int(input('Should we change max value of function argument(default=100), 1 - change, 0 - use default: '))
if x_max:
    x_max = float(input('Enter the right bound of the interval'))
else:
    x_max=100
threshold = int(input('Should we exit threshold by search interval length(default=1e-8), 1 - change, 0 - use default: '))
if threshold:
    threshold = float(input('Enter the right bound of the interval'))
else:
    threshold=1e-8
max_iter = int(input('Change maximum iteration of algorithm(default=500), 1 - change, 0 - use default: '))
if max_iter:
    int(input('Enter the right bound of the interval'))
else:
    max_iter = 500
print_info = bool(int(input('Should we print information in each iteration, 1 - yes, 0 - no: ')))
plot = bool(int(input('Should we draw a plot, 1 - yes, 0 - no: ')))
record_info = bool(int(input('Make pd.DataFrame with recorder information, 1 - yes, 0 - no: ')))
if record_info:
    f_analytic, x_res,df = optimize.bfgs(f_analytic, x0, c1, c2, x_max, threshold,  max_iter, print_info, record_info)
    print('f_analytic:', f_analytic,'x_res:', x_res,'\nres in DataFrame\n', df)
else:
    f_analytic, x_res,df = optimize.bfgs(f_analytic, x0, c1, c2, x_max, threshold,  max_iter, print_info, record_info)
    print('f_analytic:', f_analytic,'x_res:', x_res)
```
