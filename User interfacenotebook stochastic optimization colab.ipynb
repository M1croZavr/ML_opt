```python
import sympy
import numpy as np
from twovarextremas import utils_twovarextremas
from stochasticoptimization import optimize, models
```

## stoch_descent

```python
fun_anl = input('Enter function for optimization: ')
lr = float(input('Enter learning rate (float): '))
eps = float(input('Enter precision of result (float or int), default=1e-5: '))
max_iter = int(input('Enter maximum iteration of algorithm(int), default=500: '))
print_info = bool(int(input('Should we print information each iteration(1 - Yes, 0 - No), default=0: ')))
plot = bool(int(input('Should we draw plot, plot is available if less than 3 variables in function (1 - Yes, 0 - No), default=0: ')))
record_info = bool(int(input('Should we make pd.DataFrame with recorder information (1 - Yes, 0 - No), default=0: ')))
x_init_str = input('Initial x vector, (sequence of numbers, separated by space): ')
x_init = np.array([int(x) for x in x_init_str.split(' ')])
momentum = float(input('Enter momentum parameter for optimizing by gradient descent(float), default=0.85: '))

optimize.stoch_descent(fun_anl=fun_anl, lr=lr, eps=eps, max_iter=max_iter, print_info=print_info, plot=plot,
              record_info=record_info, x_init=x_init, momentum=momentum)
```

```python
fun_anl = input('Enter function for optimization: ')
fun = utils_twovarextremas.preproc_fun(fun_anl)
variables = tuple(fun.atoms(sympy.Symbol))
bounds = dict()
for var in variables:
    bound = input(f'Enter left and right bound separated by space for "{var}": ')
    bounds[str(var)] = [int(b) for b in bound.split(' ')]
t_max = float(input('Enter initial temperature (float), default=10: '))
t_min = float(input('Enter minimal temperature (float), default=0.001: '))
max_iter = int(input('Enter maximum iteration of algorithm (int), default=500: '))
plot = bool(int(input('Should we draw a plot (1 - Yes, 0 - No), deafault=0: ')))

optimize.simulated_annealing(fun_anl=fun_anl, bounds=bounds, t_max=t_max, t_min=t_min, max_iter=max_iter, plot=plot)
```

## genetic_algorithm

```python
fun_anl = input('Enter function for optimization: ')
fun = utils_twovarextremas.preproc_fun(fun_anl)
variables = tuple(fun.atoms(sympy.Symbol))
bounds = dict()
for var in variables:
    bound = input(f'Enter left and right bound separated by space for "{var}": ')
    bounds[str(var)] = [int(b) for b in bound.split(' ')]
n_bits = int(input('Enter bits per one variable (int), default=16: '))
n_pop = int(input('Enter population size (int), default = 100: '))
r_cross = float(input('Enter crossover rate (float), defualt=0.9: '))
max_iter = int(input('Enter maximum iteration of algorithm (int), defualt = 500: '))
plot = bool(int(input('Should we draw a plot (1 - Yes, 0 - No), deafault=0: ')))

optimize.genetic_algorithm(fun_anl=fun_anl, bounds=bounds, n_bits=n_bits, n_pop=n_pop, r_cross=r_cross,
                              max_iter=max_iter, plot=plot)
```
