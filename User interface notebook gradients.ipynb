## Метод градиентного спуска с постоянным шагом

```python
params = dict()
fun_anl = input('Enter the function in analytical form: f = ')
params['fun_anl'] = fun_anl
lr = float(input('Enter the constant learning rate for descent: '))
params['lr'] = lr
eps = float(input('Change precision of result(default=1e-5), 1 - change, 0 - use default: '))
if eps:
    eps = float(input('Enter the precision of result(default=1e-5): '))
    params['eps'] = eps
max_iter = int(input('Change maximum iteration of algorithm(default=500), 1 - change, 0 - use default: '))
if max_iter:
    max_iter = int(input('Enter your maximum iteration of algorithm: '))
    params['max_iter'] = max_iter
print_info = bool(int(input('Should we print information in each iteration , 1 - yes, 0 - no: ')))
params['print_info'] = print_info
record_info = bool(int(input('Make pd.DataFrame with recorder information , 1 - yes, 0 - no: ')))
params['record_info'] = record_info
plot = bool(int(input('Make plot , 1 - yes, 0 - no: ')))
params['plot'] = plot
x_init = float(input('Initialize parameters, 1 - initialize, 0 - use default: '))
if x_init:
    x_init = list(map(float, input('Enter the initial x vector, separated by space: ').split()))
    params['x_init'] = x_init
optimize.constant_descent(**params)
```
