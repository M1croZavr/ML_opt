## Метод градиентного спуска с постоянным шагом

```python
%%time
opt.constant_descent('x**4 - 5*x**2-3*x', 0.01,
                     plot=True, x_init=[0], print_info=True, record_info=True)
```

```python
%%time
opt.constant_descent('x**2 + y**2', 0.01, plot=True)
```

```python
%%time
opt.constant_descent('2*x**2 - 1.05*x**4 + (x**6 / 6) + x*y+y**2', 0.01, plot=True)
```

```python
%%time
opt.constant_descent('-cos(x)*cos(y)*exp(-((x - pi)**2 + (y-pi)**2))', 
                     0.05, print_info=False, record_info=False, plot=True, x_init=[2, 2])
```

```python
%%time
opt.constant_descent('(x+2y-7)**2+(2x+y-5)**2', 
                     0.01, print_info=False, record_info=False, plot=True)
```

Выводы:
1) У первой функции выдает правильный результат.

2) У второй функции правильный ответ. Работает корректно даже без начальной точки.

3) У третьей функции правильный ответ. Работает корректно даже без начальной точки.

4) У четвертой функции выдает правильный ответ, только при условии, что начальная точка находится близко к минимуму. Иначе может не сойтись к минимуму.

5) У пятой функции выдает правильный резульат.

Среднее время выполнения: 1,4424 секунды при условии отрисовок.
