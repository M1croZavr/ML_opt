## Метод логарифмических барьеров

```python
fun = input("Enter function: ")
g = input("Enter constraints, separated by comma: ").split(',')
x_init = np.array([int(x) for x in input("inter initial approximations, separatde by comma: ").split(',')])

if int(input("Change initialization for logarithmic barrier coefficient (default=0.1)? 1 - Yes, 0 - No: ")):
    t = float(input("Your initialization for logarithmic barrier coefficient: "))
else:
    t = 0.1

if int(input("Change multiplier of t each iteration of barrier method(default=2.)? 1 - Yes, 0 - No: ")):
    c = float(input("Your multiplier of t each iteration of barrier method: "))
else:
    c = 2.

if int(input("Change precision for logarithmic barrier part of solution(default=1e-4)? 1 - Yes, 0 - No: ")):
    tol_barrier = float(input("Your precision for logarithmic barrier part of solution: "))
else:
    tol_barrier = 1e-4

if int(input("Change precision for newton gradient method part of solution(default=1e-4)? 1 - Yes, 0 - No: ")):
    tol_newton = float(input("Your precision for newton gradient method part of solution: "))
else:
    tol_newton = 1e-4

if int(input("Change number of iterations in newton gradient method(default=50)? 1 - Yes, 0 - No: ")):
    maxiter = int(input("Your number of iterations in newton gradient method: "))
else:
    maxiter = 50

plot = bool(int(input("Plot the graph? 1 - Yes; 0 - No: ")))

log_barriers(fun=fun, g=g, x_init=x_init, t=t, c=c, tol_barrier=tol_barrier,
             tol_newton=tol_newton, maxiter=maxiter, plot=plot)
```
