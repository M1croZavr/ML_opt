## simulated_annealing

```python
fun_anl = input('Enter function for optimization: ')
fun = utils_twovarextremas.preproc_fun(fun_anl)
variables = tuple(fun.atoms(sympy.Symbol))
bounds = dict()
for var in variables:
    bound = input(f'Enter left and right bound separated by space for "{var}": ')
    bounds[str(var)] = [int(b) for b in bound.split(' ')]
optimize.simulated_annealing(fun_anl='2*x**2 - 1.05*x**4 + (x**6 / 6) + x*y+y**2', 
                             bounds=bounds, max_iter=10000, plot=1)
```

## SVM

```python
from classification import models
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.datasets import make_gaussian_quantiles
```

```python
X1, y1 = make_gaussian_quantiles(n_features=2, 
                                 n_classes=2, 
                                 random_state=2)
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=12)
```

```python
sgd_svm = models.StochasticSVM(C=0.1, eps=0.00001, lr=0.001, degree=5, plot=1)
sgd_svm.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(sgd_svm.predict(X_test), y_test))
```

## Вывод

В первую очередь хотелось бы сравнить данные методы, с методами из прошлых работ для поиска экстремума. Могу смело отметить, что скорость намного выше у стохастических. Первый метод работает идеально, особо не надо настраивать гиперпарметры для получения точного ответа, у всех функций правильно нашел точки, и главное быстро. С последним методом тоже самое. Но по второму алгоритму есть замечания, находит точки либо с большой погрешностью, либо вовсе неправильно, и выдает слишком большой разброс между точками. 

Стохастический SVM работает явно хуже чем обычный, при подборах гиперпарметров, дает максимальную оценку 0.8, и с большим разбросом.
