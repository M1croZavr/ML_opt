## Поиск локальных экстремумов функции 2-х переменных
### Используйте нижнюю ячейку

```python
vars_ = input('Enter variable names separated by spaces: ').split()
f_analytic = input('Enter the function in analytical form: f = ')
constraints = [] if input('Constraints needed 1/0: ') == '1' else None
if constraints == []:
    constraints.append(tuple(map(lambda x: float(x.strip()), input('Enter constraints on variable x separated by space: ').split())))
    constraints.append(tuple(map(lambda x: float(x.strip()), input('Enter constraints on variable y separated by space: ').split())))
plot1_flag = eval(input('Plot needed True/False: ').strip().lower().capitalize())
plot2_flag = eval(input('Level lines plot needed True/False: ').strip().lower().capitalize())
find_extrema.find_local_2var(vars_, f_analytic, constraints, plot1_flag, plot2_flag)
```

## Поиск локальных экстремумов функции двух переменных с ограничениями (метод Лагранжа)
### Используйте нижнюю ячейку

```python
vars_ = input('Enter variable names separated by spaces: ').split()
f_analytic = input('Enter the function in analytical form: f = ')
f_c = input('Enter the limiting function in analytical form: g = ')
constraints = [] if input('Constraints needed 1/0: ').strip() == '1' else None
if constraints == []:
    constraints.append(tuple(map(lambda x: int(x.strip()), input('Enter constraints on variable x separated by space: ').split())))
    constraints.append(tuple(map(lambda x: int(x.strip()), input('Enter constraints on variable y separated by space: ').split())))
plot1_flag = eval(input('Plot needed True/False: ').strip().lower().capitalize())
plot2_flag = eval(input('Level lines plot needed True/False: ').strip().lower().capitalize())
find_extrema.find_lagrange_2var(vars_, f_analytic, f_c, constraints, plot1_flag, plot2_flag)
```
