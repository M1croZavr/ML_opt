```python
from classification import models
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.datasets import make_classification
from sklearn.datasets import make_gaussian_quantiles
```

## Тестирование на датасете make classification

```python
X, y = make_classification(n_samples=100, n_features=2, 
                           n_redundant=0, n_informative=2, 
                           n_clusters_per_class=2, 
                           random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
lr_ridge = models.LogisticRegressionRidge(C=1, eps=0.00001, lr=0.001, degree=7, plot=1)
lr_ridge.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_ridge.predict(X_test), y_test))
```

```python
lr_rbf = models.LogisticRegressionRBF(C=1, eps=0.00001, gamma=0.1, plot=1)
lr_rbf.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_rbf.predict(X_test), y_test))
```

```python
lr_lasso = models.LogisticRegressionLasso(C=1, eps=0.00001, lr=0.001, degree=5, plot=1)
lr_rbf.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_rbf.predict(X_test), y_test))
```

```python
linear_svc = models.SupportVectorClassifier(C=1, eps=0.00001, kernel='poly', 
                                            degree=3, plot=1)
linear_svc.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(linear_svc.predict(X_test), y_test))
```

```python
dual_svc = models.SupportVectorClassifierDual(reg='ridge', C=1, eps=0.00001, 
                                             degree=1, plot=1)
dual_svc.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(dual_svc.predict(X_test), y_test))
```

Данный датасет был легким для моделей. На нем все модели хорошо отработали. 
Первая модель показала худший результат, но он также достаточно хорого, у всех остальный точность 1. Также, стоит отметить, что для первой модели очень важен гиперпараметр, отвечающий за степень генерации полиномиальных признаков, для других алгоритмов он не так уж и важен.

## Тестирование на датасете make_gaussian_quantiles

```python
X1, y1 = make_gaussian_quantiles(n_features=2, 
                                 n_classes=2, 
                                 random_state=2)
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=12)
```

```python
lr_ridge = models.LogisticRegressionRidge(C=1.3, eps=0.00001, lr=0.001, degree=7, plot=1)
lr_ridge.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_ridge.predict(X_test), y_test))
```

```python
lr_rbf = models.LogisticRegressionRBF(C=1, eps=0.00001, gamma=0.1, plot=1)
lr_rbf.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_rbf.predict(X_test), y_test))
```

```python
lr_lasso = models.LogisticRegressionLasso(C=1, eps=0.00001, lr=0.0001, degree=7, plot=1)
lr_rbf.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(lr_rbf.predict(X_test), y_test))
```

```python
linear_svc = models.SupportVectorClassifier(C=1, eps=0.00001, kernel='rbf', 
                                            degree=8, plot=1)
linear_svc.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(linear_svc.predict(X_test), y_test))
```

```python
dual_svc = models.SupportVectorClassifierDual(reg='ridge', C=1, eps=0.00001, 
                                             degree=7, plot=1)
dual_svc.fit(X_train, y_train)
print('Accuracy: ', metrics.accuracy_score(dual_svc.predict(X_test), y_test))
```

С этим датасетом первая модель не смогла справиться. При подборе различных гипермапарметров она не смогла показать оценку выше 0.55. Второй и третий алгоритм показали одинаковый результат. Классический Support Vector Classifier не смог показать результат выше 0.85, поэтому он проиграл остальным трем. Последний SVC показал такую же точно и график как вторая и третья модели.
