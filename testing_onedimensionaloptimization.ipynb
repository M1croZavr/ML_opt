### Из источников в интернете (указаны ниже) мы нашли одномерные унимодальные функции (указаны в массиве functions) для тестирования эффективности наших алгоритмов оптимизации.

```python
functions = ['-x**(2/3) - (1 - x**2)**(1/3)', '-(16*x**2 - 24*x +5)*exp(-x)', '(x**2 - 5*x + 6)/(x**2 + 1)', 'exp(3x)+5exp(-2x)', '-3*x*sin((3/4)*x)+exp(-2*x)']
points = [(0.0001, 0.99), (1.9, 3.9), (-5, 5), (0, 1), (0, 2 * pi), (-5.12, 5.12)]
```

```python
for i in range(len(functions)):
    print('Сечение',functions[i], optimize.golden_section_search(functions[i], points[i][0], points[i][1], record_info=False, plot=False))
    print('Параболы',functions[i], optimize.parabola_method(functions[i], points[i][0], points[i][1], record_info=False, plot=False))
    print('Брент',functions[i], optimize.brent_method(functions[i], points[i][0], points[i][1], record_info=False, plot=False))
    print('БФГШ',functions[i], optimize.bfgs(functions[i], points[i][0], record_info=False))
```

## Мы получили, что метод парабол и метод брента сходятся к правильному ответу. Метод золотого сечения и Алгоритм Бройдена — Флетчера — Гольдфарба — Шанно не дают правильного ответа. Но метод золотого сечения предполагает поиск экстремума на унимодальной одномерной функции. В нашем случае метод золотого сечения и правда нашел экстремум, но не глобальный. В случае алгоритма Бройдена — Флетчера — Гольдфарба — Шанно функция является достаточно чувствительной к значению начальную точки х0, указанный в коде. Если мы изменим значение с какой-либо границы на значение, например, 0.4, то алгоритм сойдется к 0, что нам и надо.

```python
f_res, x_res, df = optimize.bfgs('10 + x^2-10cos(2*3.14x)', 0.4, record_info=True)
```
